The objective of the project is to provide the company a cleaned and organized customer call list to send off the to call centre agents. The most important task is to provide a well-organized list for easy reading.

```python
import pandas as pd
import re
df = pd.read_excel('/content/Customer Call List.xlsx')
df
```

```python
#drop the duplicates that occur in the dataset
df = df.drop_duplicates()
df
```

```python
#remove columns that you do not need
df = df.drop(columns = 'Not_Useful_Column')
df
```

```python
#the dataset has errors in the Last_Name column (forward slash, periods, etc)
#we will need to clean this by using .strip
df['Last_Name'] = df['Last_Name'].str.strip('._/')
df
```

```python
#We need to fix the different formats of phone numbers and make it all cohesive
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)

#then turn the values into strings so we can format them
df['Phone_Number'] = df['Phone_Number'].apply(str)

#use lambda function to format the phone numbers
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3]+'-' + x[3:6]+ '-' + x[6:10])

#remove null values with the replace() function
df['Phone_Number']= df['Phone_Number'].str.replace('nan--', '')
df['Phone_Number']= df['Phone_Number'].str.replace('Na--','')
```

```python
#Formatting the addresses
df[['Street_Address', 'State','Zip_code']] = df['Address'].str.split(',', expand=True)
df
```

```python
#Replace Yes into Y and No into N
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')

#remove N/a value
df['Paying Customer'] = df['Paying Customer'].str.replace('N/a', '')
df
```

```python
#Do the same steps for the Do_Not_Contact column
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')

#convert NaN values into empty strings
df['Do_Not_Contact'] = df['Do_Not_Contact'].fillna('')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('NaN', '')
df
```

```python
#Remove customers that we cannot contact or have no phone numbers
for x in df.index:
  if df.loc[x, 'Do_Not_Contact'] == 'Y':
    df.drop(x, inplace=True)

#Remove customers that we cannot contact or have no phone numbers
for x in df.index:
  if df.loc[x, 'Phone_Number'] == '--':
    df.drop(x, inplace=True)
df
```

```python
#Reset the index numbers
df.reset_index(drop=True)
```
