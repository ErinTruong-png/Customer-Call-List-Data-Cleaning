This dataset will explore and analyze the accuracy of Punxsutawney Phil in predicting the climate during groundhog season. If the shadow is seen, there will be more weeks of winter whereas if there is no shadow, spring will come early

The dataset is from Kaggle: https://www.kaggle.com/datasets/groundhogclub/groundhog-day

**Guiding Questions:**
1. What is the average temperatures across the Northeast, Mideast, and Pennsylvania in February and March?
2. How often does Phil see his shadow?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/drive/MyDrive/Data analysis/Projects/Data files/Climate/Groundhog_climate.csv')
df
```

```python
rows = df.shape[0]
columns = df.shape[1]

print(rows,columns)
```

```python
#Check for missing data with a loop
cols_with_missing = {}
for columns in df.columns:
  missing_cols = rows - df[columns].count()
  if missing_cols:
    cols_with_missing[columns] = missing_cols
    print(columns, cols_with_missing[columns])
```

```python
#remove duplicate columns
df = df.drop_duplicates()
df
```

#**Clean the Data**

```python
#See how many NaN values there are
df.isna().sum()
```

```python
df = df.dropna()
df
```

#**Graph February average temperatures**

```python
#Graph February average temperatures by year
february = df.drop(columns = ['Punxsutawney Phil','March Average Temperature','March Average Temperature (Northeast)',
                   'March Average Temperature (Midwest)','March Average Temperature (Pennsylvania)'])
february
```

```python
plt.style.use('seaborn-v0_8-pastel')
#Height and Width of graph so we can see the dates clearly
plt.rcParams['figure.figsize'] = 25,6
# Plotting multiple lines
plt.plot(february['Year'], february['February Average Temperature'], label='Total Average')
plt.plot(february['Year'], february['February Average Temperature (Northeast)'], label='Northeast')
plt.plot(february['Year'], february['February Average Temperature (Midwest)'], label='Midwest')
plt.plot(february['Year'], february['February Average Temperature (Pennsylvania)'], label='Pennsylvania')

#other parameters
plt.legend(loc=2)
plt.title('February Average Temperature in the U.S from 1895-2016')
plt.xlabel('Years')
plt.ylabel('Temperature in Fahrenheit')
plt.xticks(rotation=90)
plt.show()
```

It seems like the coldest temperatures are at the Northeast which includes New York, Pennsylvania, and New Jersey and New England (Connecticut, Massachusetts, New Hampshire, Rhode Island, Vermont, and Maine).

The warmest temperatures are in the Midwest which includes Illinois, Indiana, Iowa, Kansas, Michigan, Minnesota, Missouri, Nebraska, North Dakota, Ohio, South Dakota, and Wisconsin.

#**Graph March Average Temperatures**

```python
#Modify dataframe to only have March values
march = df.drop(columns=['Punxsutawney Phil','February Average Temperature','February Average Temperature (Northeast)',
                         'February Average Temperature (Midwest)','February Average Temperature (Pennsylvania)'])
march
```

```python
#Graph March values
plt.style.use('seaborn-v0_8-pastel')
plt.rcParams['figure.figsize'] = 25,6
#plotting multiple lines
plt.plot(march['Year'], march['March Average Temperature'], label='Total Average')
plt.plot(march['Year'], march['March Average Temperature (Northeast)'], label='Northeast')
plt.plot(march['Year'], march['March Average Temperature (Midwest)'], label='Midwest')
plt.plot(march['Year'],march['March Average Temperature (Pennsylvania)'],label='Pennsylvania')

#other parameters
plt.title('March Average Temperature in the U.S from 1895-2016')
plt.xlabel('Years')
plt.ylabel('Temperature in Farenheit')
plt.legend(loc=2)

#rotate horizontal axis names
plt.xticks(rotation=90)
plt.show()
```

This graph corresponds to the February average temperature graph as the Northeast is the coldest whereas the midwest is the warmest. There are less dramatic spikes in the flucuations of temperature in March than in February but still follows a similar pattern nonetheless.

So it seems like Phil had almost always seen his shadow so that would mean that there should be six more weeks of winter according to the legend. The partial shadow had occured rarely (less than 20 counts) but that is not traditionally part of the legend as it does not provide a definitive yes or no answer.
